```python
import pandas as pd
import numpy as np
import os, glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
folder = 'coverage_ratios_results'
file1 = "summary_num_uavs_simulation"
df = pd.read_csv(str(folder+'/'+file1+'.txt'), delimiter = ",")
```

```python
average1 = df.groupby(['max_coverage_ratio','NUM_UAVS']).mean()
average2 = df.groupby(['NUM_UAVS','max_coverage_ratio']).mean()
average2
```

```python
plt.figure(figsize=(10, 7))

plt.title("Different coverage ratios and mission times")
plt.xlabel("Num of UAVs")
plt.ylabel("Mission time (seconds)")

plt.plot(average1.loc[90].index,average1.loc[90]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[90].index,average1.loc[90]['mission_time'],
            alpha=0.5, label='90 %', edgecolor='black')

plt.plot(average1.loc[95].index,average1.loc[95]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[95].index,average1.loc[95]['mission_time'],
            alpha=0.5, label='95 %', edgecolor='black')

plt.plot(average1.loc[97].index,average1.loc[97]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[97].index,average1.loc[97]['mission_time'],
            alpha=0.5, edgecolor='black', label='97 %')

plt.plot(average1.loc[98].index,average1.loc[98]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[98].index,average1.loc[98]['mission_time'],
            alpha=0.5, edgecolor='black', label='98 %')

plt.plot(average1.loc[99].index,average1.loc[99]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[99].index,average1.loc[99]['mission_time'],
            alpha=0.5, edgecolor='black', label='99 %')

num_uavs = [2,3,4,5,6]
plt.xticks(num_uavs)
plt.legend()
plt.savefig('coverage-ratio-uavs', dpi=300, bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.title("Different coverage ratios and mission times")
plt.xlabel("Coverage ratios (%)")
plt.ylabel("Mission time (seconds)")

plt.plot(average2.loc[2].index,average2.loc[2]['mission_time'], alpha=0.7)
plt.scatter(average2.loc[2].index,average2.loc[2]['mission_time'],
            alpha=0.75, label='2 UAVs', edgecolor='black')

plt.plot(average2.loc[3].index,average2.loc[3]['mission_time'], alpha=0.7)
plt.scatter(average2.loc[3].index,average2.loc[3]['mission_time'],
            alpha=0.75, label='3 UAVs', edgecolor='black')

plt.plot(average2.loc[4].index,average2.loc[4]['mission_time'], alpha=0.7)
plt.scatter(average2.loc[4].index,average2.loc[4]['mission_time'],
            alpha=0.75, label='4 UAVs', edgecolor='black')

plt.plot(average2.loc[5].index,average2.loc[5]['mission_time'], alpha=0.7)
plt.scatter(average2.loc[5].index,average2.loc[5]['mission_time'],
            alpha=0.75, label='5 UAVs', edgecolor='black')

plt.plot(average2.loc[6].index,average2.loc[6]['mission_time'], alpha=0.7)
plt.scatter(average2.loc[6].index,average2.loc[6]['mission_time'],
            alpha=0.75, label='6 UAVs', edgecolor='black')


range_labels = ['90','95','97','98','99']
range_values = [90,95,97,98,99]
plt.xticks(range_values, range_labels)
plt.legend()
plt.savefig('coverage-ratio-ratios', dpi=300, bbox_inches='tight', pad_inches=0.1)
plt.show()
```
