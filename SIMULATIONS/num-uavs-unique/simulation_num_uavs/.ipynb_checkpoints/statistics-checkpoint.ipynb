```python
import pandas as pd
import os, glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = glob.glob('*.txt')
df = pd.read_csv(file[0], delimiter = ",")
```

```python
df.head()
```

```python
average = df.groupby('NUM_UAVS').median()
average
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(13, 5))
fig.tight_layout(pad=4.0)

# Mission time plot
axs[0].set_title("Mission time vs Num Uavs")
axs[0].set_xlabel("Num of UAVs")
axs[0].set_ylabel("Mission time (seconds)")
axs[0].scatter(df['NUM_UAVS'], df['mission_time'], color='royalblue', alpha=1, edgecolor='black', zorder=2)
axs[0].plot(average.index, average['mission_time'], color='cornflowerblue', alpha=0.8, zorder=1)

# Avg instantaneous coverage plot
axs[1].set_title("Instantaneous area coverage")
axs[1].set_xlabel("Num of UAVs")
axs[1].set_ylabel("Instantaneous area coverage %")
axs[1].scatter(df['NUM_UAVS'], df['average_inst_cov_area'], color='darkorange', alpha=1, edgecolor='black', zorder=2)
axs[1].plot(average.index, average['average_inst_cov_area'], color='orange', alpha=0.8, zorder=1)

plt.show()
```
