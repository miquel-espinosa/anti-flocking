```python
import pandas as pd
import numpy as np
import os, glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
folder = 'communication_range'
file = glob.glob(str(folder+'/'+'*.txt'))
df = pd.read_csv(file[0], delimiter = ",")
```

```python
df.head()
```

```python
# Adding field for communication range in df
values = [15,30,45,60]
counter = 0
value = 15
for index, row in df.iterrows():
    if index%10==0 and index!=0:
        counter += 1
        j = counter%len(values)
        value = values[j]
    df.loc[index, 'comm-type'] = value
```

```python
pd.set_option('display.max_rows', 500)
df.head(130)
```

```python
average1 = df.groupby(['comm-type','NUM_UAVS']).mean()
average2 = df.groupby(['NUM_UAVS','comm-type']).mean()
average2
```

```python
plt.figure(figsize=(8, 5.5))

plt.title("Different communication ranges and mission times")
plt.xlabel("Num of UAVs")
plt.ylabel("Mission time (seconds)")

plt.plot(average1.loc[15].index,average1.loc[15]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[15].index,average1.loc[15]['mission_time'],
            alpha=0.5, label='15 meters', edgecolor='black')

plt.plot(average1.loc[30].index,average1.loc[30]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[30].index,average1.loc[30]['mission_time'],
            alpha=0.5, label='30 meters', edgecolor='black')

plt.plot(average1.loc[45].index,average1.loc[45]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[45].index,average1.loc[45]['mission_time'],
            alpha=0.5, edgecolor='black', label='45 meters')

plt.plot(average1.loc[60].index,average1.loc[60]['mission_time'], alpha=0.7)
plt.scatter(average1.loc[60].index,average1.loc[60]['mission_time'],
            alpha=0.5, edgecolor='black', label='Always communication')

num_uavs = [3,6,9]
plt.xticks(num_uavs)
plt.legend()
plt.savefig('communication-range-uavs', dpi=300, bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.title("Different communication ranges and mission times")
plt.xlabel("Communication range (meters)")
plt.ylabel("Mission time (seconds)")

plt.plot(average2.loc[3].index,average2.loc[3]['mission_time'], alpha=0.7)
plt.scatter(average2.loc[3].index,average2.loc[3]['mission_time'],
            alpha=0.75, label='3 UAVs', edgecolor='black')

plt.plot(average2.loc[6].index,average2.loc[6]['mission_time'], alpha=0.7)
plt.scatter(average2.loc[6].index,average2.loc[6]['mission_time'],
            alpha=0.75, label='6 UAVs', edgecolor='black')

plt.plot(average2.loc[9].index,average2.loc[9]['mission_time'], alpha=0.7)
plt.scatter(average2.loc[9].index,average2.loc[9]['mission_time'],
            alpha=0.75, label='9 UAVs', edgecolor='black')

range_labels = ['15','30','45','Always communication']
range_values = [15,30,45,60]
plt.xticks(range_values, range_labels)
plt.legend()
plt.savefig('communication-range-ranges', dpi=300, bbox_inches='tight', pad_inches=0.1)
plt.show()
```
